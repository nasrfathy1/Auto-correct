```python
from nltk.tokenize import word_tokenize
import nltk
import random as rnd
import re
import pandas as pd
import numpy as np
from collections import Counter
```

```python
data=[]
regex = re.compile('[^a-zA-Z0-9\s\.,-;:!?]+')
count=0
with open('/kaggle/input/shakespeare-plays/alllines.txt') as lines:
    for line in lines :
        line= regex.sub('', line)
        line=re.sub(r'[.,-;:!?]','.',line)
        data.append(line.lower().strip().replace('"',''))

        
        
    
```

```python
def tokenize_data(data):
    clean_data=[]
    for line in data:
        line=word_tokenize(line)
        words=[]
        for word in line:
            words.append(word)
        clean_data.append(['<sos>']+['<sos>']+words+['<eos>'])
    return(clean_data)
```

```python
clean_data=tokenize_data(data)
clean_data[0:5]
```

```python
def vocab_ngram(data,n):
    vocab=[]
    ngrame=[]
    for sentence in data:
        ngrame+=nltk.ngrams(sentence,n)
        for word in sentence:
            if word not in vocab:
                vocab.append(word)
    return vocab,ngrame
```

```python
vocab,unigram=vocab_ngram(clean_data,1)

_,bigram=vocab_ngram(clean_data,2)

_,trigram=vocab_ngram(clean_data,3)
```

```python
def delete(string):
    a=[string[0:idx]+string[idx+1:] for idx in range(len(string))]
    return set(a)

def swap(string):
    a=[string[0:idx]+string[idx+1]+string[idx]+string[idx+2:] for idx in range(len(string)) if (idx + 2)<=len(string)]
    return set(a)

def replace(string):
    alpha=[chr(idx) for idx in range(97,123)]
    a=[string[0:idx]+i+string[idx+1:] for idx in range(len(string)) for i in alpha if string[idx]!=i]
    
    return set(a)


def insert(string):
    alpha=[chr(idx) for idx in range(97,123)]
    a=[string[0:idx]+i+string[idx:] for idx in range((len(string)+1)) for i in alpha ]
    
    return set(a)
```

```python
def n_edit(word,n_edit,vocab=vocab):
    words={'unknow'}
    word_in_vocab=[]
    for i in range(n_edit):
        delete_word=delete(word)
        swap_word=swap(word)
        replace_word=replace(word)
        insert_word=insert(word)
        words=words|delete_word|swap_word|replace_word|insert_word
    for i in words :
        if i in vocab:
            word_in_vocab.append(i) 
        
    return word_in_vocab
```

```python
def get_correct(text,vocab,n):
    correct_sentence=''
    words=['<sos>']+['<sos>']+word_tokenize(text)
    for idx in range(2,len(words)):
        if words[idx] in vocab:

            correct_sentence+=' ' +words[idx]
        else:

            word_generate=n_edit(words[idx],n)
            prob=0
            correct=''
            count=0
            word_with_count=''
            for i in word_generate:
                if unigram[(i,)]>count:
                    count=unigram[(i,)]
                    word_with_count=i
                try:
                    numerator=trigram[(words[idx-2],words[idx-1],i)]
                    denumerator=bigram[(words[idx-2],words[idx-1])]
                    probability=numerator/denumerator
                    if prob <probability:
                        prob=probability
                        correct=i
                except:
                    pass
            if len(correct)<1:
                correct_sentence+=' ' +word_with_count
            else:
                correct_sentence+=' ' +correct
    return correct_sentence
```
